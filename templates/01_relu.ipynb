### 🟢 简单：实现 ReLU

从零开始实现 **ReLU**（Rectified Linear Unit, 修正线性单元）激活函数。

$$\text{ReLU}(x) = \max(0, x)$$

```python

    ...
```

````python
# ✏️ 在这里写你的实现

def relu(x: torch.Tensor) -> torch.Tensor:
    # pass  # 请替换此处
    '''
    torch.maximum 逐元素（Element-wise）比较两个张量，并取其中的最大值,支持高效的并行计算和广播机制（Broadcasting）。
    torch.max() 返回单个张量中中最大值,返回标量
    torch.max(x, dim=n) 返回某维度的最大值，及其索引
    '''
    return torch.maximum(x, torch.tensor(0.0))
````

````python
# 使用布尔值
def relu(x: torch.Tensor) -> torch.Tensor:
    out = x.clone()
    out = out[out < 0] # out < 0 取小于 0 的索引
    return out
````

````python
# 使用 clamp 截断
def relu(x: torch.Tensor) -> torch.Tensor:
    '''
    torch.clamp(x, min=n, max=m) 将张量限制在特定范围内(夹钳)
    '''
    return torch.clamp(x, min=0)
````
