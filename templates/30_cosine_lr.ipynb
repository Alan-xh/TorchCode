### 调度
```
step < warmup:  lr = max_lr * step / warmup_steps  (线性递增)
step >= warmup: lr = min_lr + 0.5*(max_lr-min_lr)*(1 + cos(π * progress))
```
其中 `progress = (step - warmup) / (total - warmup)`
