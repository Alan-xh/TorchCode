# 🟠 中等：LoRA（低秩适配）

实现 **LoRA** — 大模型的参数高效微调方法。

$$h = W_0 x + \frac{\alpha}{r} B A x$$

### 函数签名
```python
class LoRALinear(nn.Module):
    def __init__(self, in_features, out_features, rank, alpha=1.0): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### 要求
- `self.linear`：冻结的 `nn.Linear`（权重和偏置 `requires_grad=False`）
- `self.lora_A`：`nn.Parameter(rank, in_features)` — 随机初始化
- `self.lora_B`：`nn.Parameter(out_features, rank)` — **零**初始化
- 缩放：`alpha / rank`
