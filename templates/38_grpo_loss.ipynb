# 🔴 困难: GRPO 损失

实现**分组相对策略优化 (GRPO)** 损失——一种分组式、基线减去的 REINFORCE 目标函数，常用于 RLAIF (来自 AI 反馈的强化学习)。

给定一批 log 概率、标量奖励和分组 ID（每个提示一个分组），定义分组内标准化优势：

$$A_i = \frac{r_i - \bar r_{g(i)}}{\text{std}_{g(i)} + \epsilon}$$

其中 \(\bar r_{g(i)}\) 和 \(\text{std}_{g(i)}\) 是示例 \(i\) 所属分组的奖励均值和标准差。

GRPO 损失是负优势加权 log 概率：

$$\mathcal{L}_{\text{GRPO}} = -\mathbb{E}_i \big[\,\text{stop\_grad}(A_i)\, \log \pi_\theta(y_i)\big].$$
