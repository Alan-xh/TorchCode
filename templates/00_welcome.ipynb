## 📁 文件结构说明

每个题目都配有**两个**笔记本：

| 文件                        | 说明                          |
|-----------------------------|-------------------------------|
| `01_relu.ipynb`             | ✏️ **空白模板** — 在这里写你的实现 |

## 使用方法

1. **打开一道题目** — 选择空白模板，例如 `01_relu.ipynb`
2. **阅读题目描述** — 第一格 cell 包含问题说明和函数签名
3. **编写代码** — 在标有 `✏️ YOUR IMPLEMENTATION HERE` 的 cell 中实现
4. **调试** — 可以随意插入 cell 来 `print(x.shape)` 等
5. **提交检查** — 运行最后一个 cell `check("relu")` 查看结果
6. **卡住了？** — 可以调用 `hint("relu")` 获取提示，或者打开对应的 `_solution.ipynb`
7. **重置** — 点击工具栏中的 🔄 **Reset** 按钮可恢复空白模板重新练习

> 💡 每个笔记本都带有 **Colab** 工具栏按钮和 **Open in Colab** 徽章 — 可直接一键在 Google Colab 中运行，无需任何环境配置。
